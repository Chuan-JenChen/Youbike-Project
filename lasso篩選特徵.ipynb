```python
ds= DataFrame(Bike)
X_choice = ds.loc[:, ['sno',  'tot', 'Hr', 'weekday_num', 'holiday', 'school_off', 'Temperature', 'RH', 'WS',
       'rain', 'Precp', 'MRT_Out_ppl', 'MRT_Dist', 'School_Dist',  'Weather_Dist', 'Generation', 'UVI', 'return_sec','rent_sec', 'rent_count']]
ds = X_choice.values
X = ds[:, 0:19]
y = ds[:, 19]
y = y.astype('int64')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
model = linear_model.Lasso(alpha=0.1)
model.fit(X_train, y_train)
diabetes_y_pred = model.predict(X_test)
```

```python
# The coefficients
print('Coefficients: {}\n'.format(model.coef_))
# The mean squared error
print('number of model coef: {}'.format(np.sum(model.coef_ != 0)))
```
